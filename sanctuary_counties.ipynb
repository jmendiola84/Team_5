```python
import pandas as pd
```

```python
x = "sanctuary_cities.csv"
y = "county_fips.csv"
```

```python
xx = pd.read_csv(x)
yy = pd.read_csv(y)
```

```python
xx.head()
```

```python
yy.head()
```

```python
sanctuary_counties_df = pd.merge(xx, yy, on="County")
```

```python
sanctuary_counties_df.head()
```

```python
sanctuary_counties_df
```

```python
sanctuary_counties_us = sanctuary_counties_df.rename(columns={"State_x": "State"})
```

```python
sanctuary_counties_us.head()
```

```python
del sanctuary_counties_us['State_y']
```
