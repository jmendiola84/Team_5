```python
# Dependencies
import json
from pprint import pprint
import pandas as pd
import numpy as np
import time 
from datetime import datetime

path = "resources/EO_Immigration_Tweets_full.json"
```

```python
output_list = pd.DataFrame(columns=['created_at','account_location','text_tweet', 'place_type',
                                    'country_code', 'place_name_long', 'place_name_short' 
#                                    ,'coordinate_box1_lat','coordinate_box2_lat','coordinate_box3_lat','coordinate_box4_lat',
#                                    'coordinate_box1_lon','coordinate_box2_lon','coordinate_box3_lon','coordinate_box4_lon'
])

final_output_list = pd.DataFrame()
# "Real Person" Filters
min_tweets = 4
max_tweets = 10000
max_followers = 5000
max_following = 5000
lang = "en"
country = "US"
# Set other parameters
entry = 0
unique_ids = []
```

```python
# Super Twitter Loop    
with open(path, 'r') as f:
    entry = 0
    for line in f:
        tweet = json.loads(line)
        #pprint(tweet)
        # Use filters to check if user meets conditions
        if (tweet["user"]["followers_count"] < max_followers and
            tweet["user"]["statuses_count"] > min_tweets and
            tweet["user"]["statuses_count"] < max_tweets and
            tweet["user"]["friends_count"] < max_following and
            tweet["user"]["lang"] == lang and (
                (tweet["place"] is not None and
                 tweet["place"]['country_code'] is country) or
                tweet['user']["location"] is not ''
            # or tweet['user']["location"] is not None
            )): 
                # Loop through all tweets
                tweet_id = tweet_id = tweet["id"]
                #Read information
                created_at= (tweet['created_at'])
                created_at =datetime.strptime(created_at, "%a %b %d %H:%M:%S %z %Y")
                account_location = tweet['user']["location"]
                text_tweet = tweet["full_text"]
                my_place = tweet["place"]
                if my_place is None: 
                    place_name_short = np.nan
                    place_name_long = np.nan
                    place_type = np.nan
                    country_code = np.nan
#                     coordinate_box1_lat = np.nan
#                     coordinate_box2_lat = np.nan
#                     coordinate_box3_lat = np.nan
#                     coordinate_box4_lat = np.nan
#                     coordinate_box1_lon = np.nan
#                     coordinate_box2_lon = np.nan
#                     coordinate_box3_lon = np.nan
#                     coordinate_box4_lon = np.nan
                else:
                    place_name_short = tweet["place"]['name']
                    place_name_long = tweet["place"]['full_name']
                    place_type = tweet["place"]['place_type']
                    country_code = tweet["place"]['country_code']
                    #pprint(tweet["place"]['country_code'])
                output_list.at[entry, 'created_at'] = created_at
                output_list.at[entry, 'text_tweet'] = text_tweet
                output_list.at[entry, 'account_location'] = account_location
                output_list.at[entry, 'place_name_short'] = place_name_short         
                output_list.at[entry, 'place_name_long'] = place_name_long         
                output_list.at[entry, 'place_type'] = place_type         
                entry += 1
                if entry % 1000 == 0:
                    print(entry)
```
